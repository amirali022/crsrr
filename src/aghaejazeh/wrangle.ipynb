```python
import numpy as np
import pandas as pd
import os
```

```python
list_path = "../../data/aghaejazeh/tutorsList/";

tutors_lists = np.array( os.listdir( list_path))

tutors_lists = tutors_lists[ np.flatnonzero( np.core.defchararray.find( tutors_lists, "csv") != -1)]

tutors_list = tutors_lists[ -1]

tl_df = pd.read_csv( f"{ list_path}/{ tutors_list}")

tl_df.head()
```

```python
detail_path = "../../data/aghaejazeh/tutorsDetail/";

details_lists = np.array( os.listdir( detail_path))

details_lists = details_lists[ np.flatnonzero( np.core.defchararray.find( details_lists, "csv") != -1)]

details_list = details_lists[ -1]

td_df = pd.read_csv( f"{ detail_path}/{ details_list}")

td_df.head()
```

```python
df = tl_df.merge( td_df, how="inner", left_on="link", right_on="url")

df.head()
```
