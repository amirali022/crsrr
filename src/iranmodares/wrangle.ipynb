```python
import os
import numpy as np
import pandas as pd
```

```python
tutors_list_path = "../../data/iranmodares/tutorList"

tutors_list_datasets = np.array( os.listdir( tutors_list_path))

tutors_list_datasets = tutors_list_datasets[ np.flatnonzero( np.core.defchararray.find( tutors_list_datasets, "csv") != -1)]

tutors_list_dataset = tutors_list_datasets[ -1]

tl_df = pd.read_csv( f"{ tutors_list_path}/{ tutors_list_dataset}")

tl_df.head()
```

```python
path = "../../data/iranmodares/tutorDetail"

listDatasets = np.array( os.listdir( path))

listDatasets = listDatasets[ np.flatnonzero( np.core.defchararray.find( listDatasets, "csv") != -1)]

print( listDatasets)
```

```python
dfs = []

for ds in listDatasets:
	dfs.append( pd.read_csv( f"{ path}/{ ds}"))

df = pd.concat( dfs)

df.head()
```

```python
merged = pd.merge( df, tl_df, left_on="id", right_on="id", how="inner")

merged.head()
```

```python
cleaned = merged.dropna( subset=[ "phone_x", "phone_y"], how="all")
```

```python
unique = cleaned.drop_duplicates( subset=[ "phone_x", "phone_y", "email"])
```

```python
unique.to_csv( "output.csv")
```
